```python
# Creating classifications of Profotabilility by dividing records basis the number of Orders created in a day

def label_profitability_orders (row):
    if row['Orders_y'] > 20000 :
        return '3'
    elif row['Orders_y'] > 15000 :
        return '2'
    elif row['Orders_y'] > 10000 :
        return '1'
    return '0'
```

```python
df_retail_store_order['profitability_orders'] = df_retail_store_order.apply (lambda row: label_profitability_orders(row), axis=1)
```

```python
df_retail_store_order
```

```python
# libraries for modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
from sklearn.tree import export_graphviz
from IPython.display import Image
import graphviz
```

```python
# Creating a copy of the data set for first round of modelling
df1 = df_retail_store_order.copy()
```

```python
# dropping null valules
df1.isnull().sum()
```

```python
# dropping features without values for this model
df1.drop(columns=['StoreType', 'Assortment', 'CompetitionDistance', 'CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2SinceWeek', 'Promo2SinceYear', 'PromoInterval' ], axis=1, inplace= True)
```

```python
df1.drop(columns=['Id', 'Store', 'Orders_y' ], axis=1, inplace= True)
```

```python
df1.isnull().sum()
```

```python
df1.nunique(axis=0)
```

```python
df1.describe()
```

```python
def encoding_stateHoliday (row):
    #Assuming 0 is no holiday, any other value is a hoiday
    if row['StateHoliday'] != '0' :
        return 1
    return 0

df1['StateHoliday'] = df1.apply (lambda row: encoding_stateHoliday(row), axis=1)
```

```python
df1
```

```python
X = df1.drop('profitability_orders', axis=1)
y = df1['profitability_orders']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_prediction = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_prediction)
print("Random Forest, Model Accuracy:", accuracy)
```

```python
y_prediction_trainSet = rf.predict(X_train)
accuracy_trainSet = accuracy_score(y_train, y_prediction_trainSet)
print("Random Forest, Model Accuracy (trainSet):", accuracy_trainSet)
```

Model Doesnt seem to be overfitting
